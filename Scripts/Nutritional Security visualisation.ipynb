```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("C:\\Users\\defaultuser0\\Downloads\\FAOSTAT_data_en_6-27-2024 new.xlsx")
```

```python
# Average dietary energy requirements

df_1 = df[df['Item'] == 'Average dietary energy requirement (kcal/cap/day)']

yearly_sum = df_1.groupby('Year')['Value'].mean().reset_index()

yearly_sum.columns = ['Year', 'Total Value']

plt.figure(figsize=(10, 6))
plt.plot(yearly_sum['Year'], yearly_sum['Total Value'], marker='o', linestyle='-', color='b')
plt.title('Yearly average dietary energy requirements of all countries')
plt.grid(True)

# Set xticks to show only alternative years and rotate them by 45 degrees
years = yearly_sum['Year']
plt.xticks(ticks=years[::2], labels=years[::2], rotation=45)

plt.tight_layout()  # Adjust layout to ensure everything fits without overlap
plt.show()
```

```python
# Undernourishment

df_1 = df[df['Item'] == 'Prevalence of undernourishment (percent) (3-year average)']

df_1['Value'] = df_1['Value'].replace('<2.5', 2.5).astype(float)

df_1 = df_1.dropna()

yearly_sum = df_1.groupby('Year')['Value'].mean().reset_index()

yearly_sum.columns = ['Year', 'Total Value']

plt.figure(figsize=(10, 6))
plt.plot(yearly_sum['Year'], yearly_sum['Total Value'], marker='o', linestyle='-', color='b')
plt.title('Yearly average of undernourishment of all countries (percent)')
plt.grid(True)

# Set xticks to show only alternative years and rotate them by 45 degrees
years = yearly_sum['Year']
plt.xticks(ticks=years[::2], labels=years[::2], rotation=45)

plt.tight_layout()  # Adjust layout to ensure everything fits without overlap
plt.show()
```

```python
df_1 = df[df['Item'] == 'Average dietary energy requirement (kcal/cap/day)']

# Assuming df is your DataFrame
# Filter for the years 2001, 2010, and 2020
df_filtered = df_1[df_1['Year'].isin([2001, 2010, 2020])]

# Plot KDEs for the specified years
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df_filtered[df_filtered['Year'] == 2001]['Value'], label='2001', fill=True)
sns.kdeplot(data=df_filtered[df_filtered['Year'] == 2010]['Value'], label='2010', fill=True)
sns.kdeplot(data=df_filtered[df_filtered['Year'] == 2020]['Value'], label='2020', fill=True)

plt.title('KDE of Average Dietary Energy Requirement for 2001, 2010, and 2020')
plt.xlabel('Average Dietary Energy Requirement (kcal/cap/day)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
